```python
# trying to compute the Fibonacci recurlsively

def fib_recursive(n):
    print(n)
    if n==0:
        return 0
    elif n==1:
        return 1
    
    else:
        return fib_recursive(n-1) + fib_recursive(n-1)
```

```python
fib_recursive(8)
```

```python
# see the number of repititions in the above functions

# lets make use of the lRU cache now

from functools import lru_cache

@lru_cache(maxsize=None)
def fib_recursive(n):
    print(n)
    if n==0:
        return 0
    elif n==1:
        return 1
    
    else:
        return fib_recursive(n-1) + fib_recursive(n-1)
```

```python
fib_recursive(8)
```
